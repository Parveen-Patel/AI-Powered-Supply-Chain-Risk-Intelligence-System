### Loading Model Predictions

This notebook analyzes shipment delay predictions generated by the machine learning model.
These predictions are stored in the Gold layer of the Lakehouse architecture.

```python
predictions_df = spark.table("supply_chain_ai.gold.shipping_delay_predictions")

display(predictions_df)
```

### Shipment Mode Sensitivity to Weather

This analysis identifies which shipment modes are most affected by weather severity.

```python
shipment_weather_risk = spark.sql("""
SELECT
    Mode_of_Shipment_index,
    weather_severity_index,
    AVG(prediction) AS avg_delay_risk,
    COUNT(*) AS total_shipments
FROM supply_chain_ai.gold.shipping_delay_predictions
GROUP BY Mode_of_Shipment_index, weather_severity_index
ORDER BY avg_delay_risk DESC
""")

display(shipment_weather_risk)
```

### Product Category Disruption Risk

This analysis evaluates which product importance categories
experience the highest delay probability.

```python
product_disruption = spark.sql("""
SELECT
    Product_importance_index,
    AVG(prediction) AS disruption_risk,
    COUNT(*) AS total_shipments
FROM supply_chain_ai.gold.shipping_delay_predictions
GROUP BY Product_importance_index
ORDER BY disruption_risk DESC
""")

display(product_disruption)
```

### Warehouse Delay Risk Analysis

This identifies warehouses where shipment delays are most likely.

```python
warehouse_risk = spark.sql("""
SELECT
    Warehouse_block_index,
    AVG(prediction) AS avg_delay_probability,
    COUNT(*) AS total_shipments
FROM supply_chain_ai.gold.shipping_delay_predictions
GROUP BY Warehouse_block_index
ORDER BY avg_delay_probability DESC
""")

display(warehouse_risk)
```

### AI Disruption Alert Engine

Shipments with predicted delay probability above 70% are flagged
as high-risk shipments. This acts as an early warning system
for supply chain disruptions.

```python
spark.sql("""
CREATE OR REPLACE TABLE supply_chain_ai.gold.high_risk_shipments AS
SELECT *
FROM supply_chain_ai.gold.shipping_delay_predictions
WHERE prediction >= 0.7
""")
```

```python
high_risk_df = spark.table("supply_chain_ai.gold.high_risk_shipments")

display(high_risk_df)
```

### Supply Chain Risk Intelligence Dataset

This aggregated dataset powers dashboards for monitoring
supply chain risk across warehouses, shipment modes,
product categories, and weather conditions.

```python
dashboard_dataset = spark.sql("""
SELECT
    Warehouse_block_index,
    Mode_of_Shipment_index,
    Product_importance_index,
    weather_severity_index,
    AVG(prediction) AS avg_delay_risk,
    COUNT(*) AS shipment_volume
FROM supply_chain_ai.gold.shipping_delay_predictions
GROUP BY
    Warehouse_block_index,
    Mode_of_Shipment_index,
    Product_importance_index,
    weather_severity_index
""")
```

```python
dashboard_dataset.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("supply_chain_ai.gold.supply_chain_risk_dashboard")

spark.sql("USE supply_chain_ai.gold")

display(spark.table("supply_chain_ai.gold.supply_chain_risk_dashboard"))
```
